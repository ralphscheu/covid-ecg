```python
import pandas as pd
```

```python
df = pd.read_csv('post_covid_patients.csv', sep=";")

df['birth_date'] = pd.to_datetime(df['birth_date'])
df['diagnose_date'] = pd.to_datetime(df['diagnose_date'])

df['age'] = (df['diagnose_date'] - df['birth_date']).astype('<m8[Y]')


df.head(15)
```

```python
df['gender'].value_counts()
```

```python
print(f'Age max: {df["age"].max()}')
print(f'Age min: {df["age"].min()}')
print(f'Age mean: {df["age"].mean():2.2f}')
```
